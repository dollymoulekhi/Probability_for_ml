# K-S Test

### Used for Checking the similarity of Two-Distribution

```python
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
```

## Generate a Gaussian Random Variable x

```python
X=stats.norm.rvs(size=1000)
sns.set_style('whitegrid')
sns.kdeplot(np.array(X),bw=0.5)
plt.show()
```

```python
result=stats.kstest(X,'norm')
print(result)
```

**pvalue  is .8426 which is representing this plot is similar to gaussian distribution**

```python
y=np.random.uniform(0,1,1000)
sns.kdeplot(np.array(y),bw=0.1)
plt.show()
```

```python
result=stats.kstest(X,y)
print(result)
```

**so the pvalue specify that X and y distributions are no at all similar**
